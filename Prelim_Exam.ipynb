```
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
import matplotlib 
import seaborn as sns
```

```
def month_profit_trace(profit, reach, make_figs=True):                     
    if (profit.shape == (4,)) and (reach.shape == (4,)): 
        w1 = np.array([reach[0], profit[0]])                              ### Vectorizing the data
        w2 = np.array([reach[1], profit[1]])
        w3 = np.array([reach[2], profit[2]])
        w4 = np.array([reach[3], profit[3]])
        
        month_total = np.add(w4, np.add(w3, np.add(w1, w2)))        ## computing the total weeks and the resultant vectors 
        month_status = np.sqrt((month_total[0]**2)+(month_total[1]**2))     ## computation for the displacement 
        alpha = 10** -6
        slope = np.arctan(month_total[1]/(alpha + month_total[0]))      ## solution for angle displacement
        print(month_total)

        plt.figure(figsize=(16,5))
        plt.title('Bebang\'s Month Post Efficiency')                          
        plt.xlim(0,1.01*np.sum(reach))                                                                               ##  The min and max value of x-axis
        plt.ylim(-np.sum(np.abs(profit)),np.sum(np.abs(profit)))                                        ## The min and max value of y_axis
        plt.xlabel('FB Post Reach Increment')                                
        plt.ylabel('Profit')
        plt.grid()
        n = 2
        
        plt.quiver(0,0, w1[0], w1[1],                                                                       ## Plotting of week 1
                   angles='xy', scale_units='xy',scale=1, color='orange', width=0.0025,    
                   label='Week 1: {:.2f}'.format(np.linalg.norm(w1)))
        
        ## put necessary vector plotting code here
        
        plt.quiver(w1[0], w1[1], w2[0], w2[1],                                              ## Plotting of week 2
        angles='xy', scale_units='xy',scale=1, color='red', width=0.005,
        label='Week 2: {:.2f}'.format(np.linalg.norm(w2)))
        plt.quiver(np.add(w1[0],w2[0]), np.add(w1[1],w2[1]),                       ## Plotting of week 3
        w3[0], w3[1], angles='xy', scale_units='xy',scale=1, color='blue', width=0.0025,
        label='Week 3: {:.2f}'.format(np.linalg.norm(w3)))
        x= np.add(w1[0],np.add(w2[0],w3[0]))                                                 ## Beginning point of week 4 
        y= np.add(w1[1],np.add(w2[1],w3[1]))
        plt.quiver(x, y,                                                                                                     ## Plotting of week 4
        w4[0], w4[1], angles='xy', scale_units='xy',scale=1, color='cyan', width=0.0025,
        label='Week 4: {:.2f}'.format(np.linalg.norm(w4)))
        plt.quiver(0,0, month_total[0], month_total[1],                                                       ##  efficiency plot
        angles='xy', scale_units='xy',scale=1, color='green', width=0.005,
        label='Efficiency: {:.2f} @ {:.2f}'.format(month_status, slope))


        plt.legend(loc='upper right')                                                                                 ## legend

        if make_figs:
            plt.savefig(f'LinAlg-Lab2-Bebang Post Eff-{int(month_status)}@{int(slope)}.png', dpi=300)

        plt.show()
    
    else:
        print('Dimension error') 
```

```
profit= np.array([-18000, 3000, 12000, 10000])    ## You can change the values to make different scenarios
reach = np.array([1000, 100, 500, 10])             ## You can change the values to make different scenarios

month_profit_trace(profit, reach, make_figs=False)
```
